정리해보자면,

혼자 탑승한 승객 보다 같이 탑승한 승객의 생존률이 더 높다. 단, 5명 이상으로 가면 생존률이 급격히 떨어진다.

탑승 위치에 따라서는 딱히 생존 여부에 관계가 보이지는 않는다.

데크의 f, g의 생존률을 토대로 f, g 부근에 손상이 가해진 걸 예상해 볼 수 있다.

가장 뚜렷하게 보여지는 데이터는 여성과 아이들의 생존률이 급격히 높으며, 남성들의 생존률은 상대적으로 매우 저조하다. 남성의 희생으로 아이와 여성이 많이 생존했다는 걸 추리할 수 있다.
